## Notebook Examples of Scraping Baseball Reference

```python
import br
import pandas as pd
```

```python
# Matt Carpenter minor and major league stats
url = "http://www.baseball-reference.com/register/player.fcgi?id=carpen002mat"
d = br.tables(url)
```

The function returns a dictionary that contains all of the sortable tables on the page. These can be turned into dataframes using the pattern below. The key is the caption of the table desired. The first element in the key value pair is a list of headers and the second is the table data.

```python
pd.DataFrame(d['Register Batting'][1], columns=d['Register Batting'][0])
```

It works for a variety of page types.
